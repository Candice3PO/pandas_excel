## We import numpy as np

```python
import numpy as np
```

## We test numpy 

```python
x = np.linspace(1,10,10)
y = np.linspace(1,20,10)
print('This is the output of linspace, x: ', x)
print('This is the output of linspace, y: ', y)
```

## We test pyplot

```python
#we import pyplot
import matplotlib.pyplot as plt
#this is just a test blablabla
```

```python
plt.plot(x,y, 'g')
plt.ylabel('y linspace')
plt.xlabel('x linspace')
plt.show()
```
